# Get names of survived people

```python
import numpy as np
import pandas as pd
from os import path
import regex
import re
```

Now let's try to match the names of survived people (from https://www.encyclopedia-titanica.org/titanic-survivors/) and the names from data set. First let's get the names of survived people.

```python
survived_data_dir = './special_data'
raw_survived = ''
with open(path.join(survived_data_dir, 'encyclopedia-titanica.html'), 'r', encoding='UTF-8') as f:
    raw_survived = f.read()
```

```python
last_names = regex.findall('<span itemprop="familyName">.*\,', raw_survived)
last_names = [str.replace(last_name, '<span itemprop="familyName">', '') for last_name in last_names]
last_names = [str.replace(last_name, '</span>,', '').capitalize() for last_name in last_names]
first_names = regex.findall('<span itemprop="givenName">.*<\/span>', raw_survived)
first_names = [str.replace(first_name, '<span itemprop="givenName">', '') for first_name in first_names]
first_names = [str.replace(first_name, '</span>', '') for first_name in first_names]
titles = regex.findall('<span itemprop="honorificPrefix">.*<\/span>', raw_survived)
titles = [str.replace(title, '<span itemprop="honorificPrefix">', '') for title in titles]
titles = [str.replace(title, '</span>', '') for title in titles]

titles_firsts = [' '.join(title_first) for title_first in zip(titles, first_names)]
names = [', '.join(first_last) for first_last in zip(last_names, titles_firsts)]
names
```
