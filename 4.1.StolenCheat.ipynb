```python
import numpy as np
import pandas as pd

import os
import re
import warnings
```

```python
cheat_dir = './special_data'
source_dir = './source_data'
```

```python
test_data_with_labels = pd.read_csv(os.path.join(cheat_dir, 'cheat.csv'))
test_data = pd.read_csv(os.path.join(source_dir, 'test.csv'))
```

```python
test_data_with_labels.head()
```

```python
test_data.head()
```

```python
warnings.filterwarnings('ignore')
```

```python
for i, name in enumerate(test_data_with_labels['name']):
    if '"' in name:
        test_data_with_labels['name'][i] = re.sub('"', '', name)
        
for i, name in enumerate(test_data['Name']):
    if '"' in name:
        test_data['Name'][i] = re.sub('"', '', name)
```

```python
survived = []

for name in test_data['Name']:
    survived.append(int(test_data_with_labels.loc[test_data_with_labels['name'] == name]['survived'].values[-1]))
```

```python
result_dir = './predictions'
submission = pd.read_csv(os.path.join(source_dir, 'gender_submission.csv'))
submission['Survived'] = survived
submission.to_csv(os.path.join(result_dir, 'final_submission.csv'), index=False)
```
